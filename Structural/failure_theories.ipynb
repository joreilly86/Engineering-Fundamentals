```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Yield strength for demonstration
yield_strength = 100

# Von Mises Failure Envelope
theta_vm = np.linspace(0, 2 * np.pi, 100)
sigma1_vm = yield_strength / np.sqrt(3) * np.cos(theta_vm)
sigma2_vm = yield_strength / np.sqrt(3) * np.sin(theta_vm)

# Tresca Failure Envelope
theta_tresca = np.array([0, np.pi/3, 2*np.pi/3, np.pi, 4*np.pi/3, 5*np.pi/3, 2*np.pi])
sigma1_tresca = yield_strength * np.cos(theta_tresca)
sigma2_tresca = yield_strength * np.sin(theta_tresca)

# Plotting
plt.figure(figsize=(10, 8))
plt.plot(sigma1_vm, sigma2_vm, label='Von Mises Failure Envelope', color='blue')
plt.plot(sigma1_tresca, sigma2_tresca, label='Tresca Failure Envelope', linestyle='--', color='green')
plt.scatter(0, 0, color='red', label='Origin (No Stress)')

# Annotations for clarity
plt.annotate('Yield Strength', xy=(yield_strength/np.sqrt(3), 0), xytext=(yield_strength/2, yield_strength/4),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )

plt.xlabel('$\sigma_1$ (Principal Stress 1)')
plt.ylabel('$\sigma_2$ (Principal Stress 2)')
plt.title('Failure Envelopes for Tresca and Von Mises Criteria')
plt.legend()
plt.axis('equal')
plt.xlim(-yield_strength, yield_strength)
plt.ylim(-yield_strength, yield_strength)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Yield strength for Tresca and von Mises
Y = 100

# Generating the von Mises ellipse
theta = np.linspace(0, 2 * np.pi, 100)
vm_radius = Y / np.sqrt(2)  # Adjusting the radius for von Mises
vm_x = vm_radius * np.cos(theta)
vm_y = vm_radius * np.sin(theta)

# Adjusting Tresca hexagon points to show as a polygon (simplified for visualization)
tresca_points = np.array([
    [Y, 0],
    [Y/2, Y*np.sqrt(3)/2],
    [-Y/2, Y*np.sqrt(3)/2],
    [-Y, 0],
    [-Y/2, -Y*np.sqrt(3)/2],
    [Y/2, -Y*np.sqrt(3)/2],
    [Y, 0]  # Closing the polygon
])

# Plotting
plt.figure(figsize=(8, 8))
plt.plot(vm_x, vm_y, label='Von Mises (Ellipse)')
plt.plot(tresca_points[:, 0], tresca_points[:, 1], 'g--', label='Tresca (Hexagon)', linewidth=2)

plt.title('Failure Criteria Comparison')
plt.xlabel('$\sigma_1$')
plt.ylabel('$\sigma_2$')
plt.legend()
plt.grid(True)
plt.axis('equal')  # Ensuring equal scaling on both axes

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Yield strength (assuming isotropic material for simplicity)
yield_strength = 100

# Von Mises ellipsoid data
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x_vm = yield_strength / np.sqrt(2) * np.outer(np.cos(u), np.sin(v))
y_vm = yield_strength / np.sqrt(2) * np.outer(np.sin(u), np.sin(v))
z_vm = yield_strength / np.sqrt(2) * np.outer(np.ones(np.size(u)), np.cos(v))

# Tresca cylinder data (simplified representation)
z_tresca = np.linspace(-yield_strength, yield_strength, 100)
theta_tresca = np.linspace(0, 2 * np.pi, 100)
theta_tresca_grid, z_tresca_grid = np.meshgrid(theta_tresca, z_tresca)
x_tresca = yield_strength * np.cos(theta_tresca_grid)
y_tresca = yield_strength * np.sin(theta_tresca_grid)

# Creating a figure with 3 subplots
fig = plt.figure(figsize=(18, 6))

# Plot Tresca
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(x_tresca, y_tresca, z_tresca_grid, color='g', alpha=0.5)
ax1.set_title('Tresca Failure Envelope')
ax1.set_xlabel('$\sigma_1$')
ax1.set_ylabel('$\sigma_2$')
ax1.set_zlabel('$\sigma_3$')

# Plot von Mises
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(x_vm, y_vm, z_vm, color='b', alpha=0.5)
ax2.set_title('Von Mises Failure Envelope')
ax2.set_xlabel('$\sigma_1$')
ax2.set_ylabel('$\sigma_2$')
ax2.set_zlabel('$\sigma_3$')

# Plot Both
ax3 = fig.add_subplot(133, projection='3d')
ax3.plot_surface(x_vm, y_vm, z_vm, color='b', alpha=0.5, label='Von Mises')
ax3.plot_surface(x_tresca, y_tresca, z_tresca_grid, color='g', alpha=0.5, label='Tresca')
ax3.set_title('Both Failure Envelopes')
ax3.set_xlabel('$\sigma_1$')
ax3.set_ylabel('$\sigma_2$')
ax3.set_zlabel('$\sigma_3$')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sphere parameters
radius = 100  # Example radius

# Meshgrid for the angles
theta = np.linspace(0, np.pi, 100)  # Polar angle
phi = np.linspace(0, 2 * np.pi, 100)  # Azimuthal angle

# Create a 2D grid of angles
theta, phi = np.meshgrid(theta, phi)

# Spherical coordinates
x = radius * np.sin(theta) * np.cos(phi)
y = radius * np.sin(theta) * np.sin(phi)
z = radius * np.cos(theta)

# Plotting
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the sphere
ax.plot_surface(x, y, z, color='b', alpha=0.5)

ax.set_xlabel('$\sigma_1$')
ax.set_ylabel('$\sigma_2$')
ax.set_zlabel('$\sigma_3$')
ax.set_title('Spherical Shape Representation')

plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Sphere parameters
radius = 100  # Example radius

# Meshgrid for the angles
theta = np.linspace(0, np.pi, 100)  # Polar angle
phi = np.linspace(0, 2 * np.pi, 100)  # Azimuthal angle

# Create a 2D grid of angles
theta, phi = np.meshgrid(theta, phi)

# Spherical coordinates
x = radius * np.sin(theta) * np.cos(phi)
y = radius * np.sin(theta) * np.sin(phi)
z = radius * np.cos(theta)

# Create the plot
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z, colorscale='Blues')])

# Update layout for a better view
fig.update_layout(title='3D Sphere in Plotly', autosize=False,
                  width=800, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))

# Show the plot
fig.show()
```
