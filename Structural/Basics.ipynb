```python
# Intro Text Cell Markdown
"""
Exploring Newton's Second Law: F = m * a
We calculate force (F) as mass (m) in kilograms times acceleration (a) in meters per second squared. The result is in Newtons.
"""

# Code Cell 1: Define Variables & Function
m = 5  # mass in kilograms
a = 2  # acceleration in meters per second squared

def calculate_force(mass, acceleration):
    return mass * acceleration

# Code Cell 2: Calculate & Print
force = calculate_force(m, a)
print("The resulting force is:", force, "Newtons")

# Visualization Cell 1: Bar Chart
import matplotlib.pyplot as plt
import seaborn as sns

# Setting Seaborn style (optional)
sns.set()

# Data for bar chart
categories = ['Mass', 'Acceleration', 'Force']
values = [m, a, force]

plt.figure(figsize=(10, 6))
plt.bar(categories, values, color=['blue', 'orange', 'green'])
plt.title('Force Calculation')
plt.ylabel('Values')
plt.show()

# Visualization Cell 2: Line Plot (Mass vs. Force)
a_values = [1, 2.5, 4, 5.5]
forces = [calculate_force(m, a_val) for a_val in a_values]

plt.figure(figsize=(10, 6))
plt.plot(a_values, forces, marker='o')
plt.title('Force vs. Acceleration')
plt.xlabel('Acceleration (m/s^2)')
plt.ylabel('Force (Newtons)')
plt.grid(True)
plt.show()
```
